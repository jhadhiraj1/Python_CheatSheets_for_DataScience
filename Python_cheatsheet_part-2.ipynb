# By Label

```python
import pandas as pd
data={"day":[0,1,2,3,4,5],
       "visitors": [1000,2000,3000,2500,1500,1300],
        "Bounce_rate": [100,200,150,135,160,101]}
df=pd.DataFrame(data,columns=['day','visitors','Bounce_rate'],index=[1,2,3,4,5,6])   ## Coulumns values should be same as Data Variables
df=pd.DataFrame(data,index=[1,2,3,4,5,6])  ## Same as above
df.loc[1]          ## Index is the key to retrieve data from DataFrame
df.loc[1][2]
df.loc[1]['visitors']
df.at[1,'Bounce_rate']      ## Same as above just remember the Syntax
#df.ix[2]         ## Select Single row of subset rows
#df.ix[1,'visitors']    ## Selecting row 1 of index from visitors

##Boolean Indexing
s=pd.Series([1,2,3])
s[s>2]   ## values>2
s[~(s>1)]    # Negation
s[(s<-1) | (s>2)]
df[df['day']>2]
s[3]=4
s
```

```python
## Dropping
import pandas as pd
data={"day":[0,1,2,3,4,5],
       "visitors": [1000,2000,3000,2500,1500,1300],
        "Bounce_rate": [100,200,150,135,160,101]}
df=pd.DataFrame(data,columns=['day','visitors','Bounce_rate'],index=[1,2,3,4,5,6]) 
df.drop(columns='Bounce_rate')
df.sort_index()
df.sort_values(by='Bounce_rate')
df.rank()
df.shape
df.index
df.columns
df.info()
df.count()  ##count the values
df.sum()       ## Sums all the entries in cols 
df.mean()
df.median()
df.std()
df.cumsum()  ## Cummulative Summation
df.min()/df.max()   ### Applies on entries by cols
print(df.idxmin()/df.idxmax())      ##forindex Values
```

```python
s=pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])
s1=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])  ##Index must be Same
s.add(s1,fill_value=2)  ## Nan in s1 replaced by 2
s.sub(s1,fill_value=3)
s.mul(s1,fill_value=6)
s.div(s1,fill_value=7)
```
